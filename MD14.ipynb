```python
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('Grid_Select_Delta.csv')
```

```python
df['Delta'].plot.hist(figsize=(15,10), color='green', title="Model Residuals")
```

```python
print('Mean = %f, Standard_deviation = %f' %(np.mean(df['Delta']), np.std(df['Delta'])))
```

```python
from sklearn import linear_model
```

```python
lm = linear_model.LinearRegression()
lm.fit(X=(df['Obs_Count'].values).reshape(-1,1) , y=(df['Pred_Count'].values).reshape(-1,1))
print('R-squared of the Predicted vs. Observed Linear Model (explained variance) = %f4' 
      % lm.score(X=(df['Obs_Count'].values).reshape(-1,1) , y=(df['Pred_Count'].values).reshape(-1,1)))
```
